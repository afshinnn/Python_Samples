** 40) Continuing with data from above, convert the dataframe content to be the right "data type" (i.e., int, float, etc) in order to perform statistical calculations for the DataFrame. Provide your code and the resulting new print out as the answer.

```python
x = df.columns
df[x] = df[x].replace({'\$':''}, regex = True)
df[x] = df[x].replace({'%':''}, regex = True)
df.head()
```

```python
# now we convert string columns to int or float
df['Avg Spending Per Episode (Hospital)'] = pd.to_numeric(df['Avg Spending Per Episode (Hospital)'], errors='coerce')
df['Avg Spending Per Episode (State)'] = pd.to_numeric(df['Avg Spending Per Episode (State)'], errors='coerce')
df['Avg Spending Per Episode (Nation)'] = pd.to_numeric(df['Avg Spending Per Episode (Nation)'], errors='coerce')
df['Percent of Spending (Hospital)'] = pd.to_numeric(df['Percent of Spending (Hospital)'], errors='coerce')
df['Percent of Spending (State)'] = pd.to_numeric(df['Percent of Spending (State)'], errors='coerce')
df['Percent of Spending (Nation)'] = pd.to_numeric(df['Percent of Spending (Nation)'], errors='coerce')
```

**43) Using the dataframe from above (Q 42), find the average spending for each different TX hospitals groupedby each kind of different claim types. 
Provide your code and the resulting new print out as the answer.**

```python
# (1) project the columns that are reuired first
# (2) remove rows with claim type 'Total' as this rows include the total Avg spending, so they are not data, they are summaries
# and we don't want them in our groupby since they can polute the summaries that we are going the obtain
# (3) group by hospital and claim type
grpby = df_tx[['Hospital', 'Claim Type', 'Avg Spending Hospital', 'Percent Spending Hospital']][df_tx['Claim Type'] != 'Total'].groupby(['Hospital', 'Claim Type'])
```

```python
grpby['Hospital', 'Claim Type', 'Avg Spending Hospital'].sum()
```

**44) Continuing from above (Q 43), what is the mean and max percentage spending of different TX hospitals for each different claim types.

```python
# mean percentage spending of different TX hospitals for each different claim types
grpby['Hospital', 'Claim Type', 'Percent Spending Hospital'].mean()
```

```python
# max percentage spending of different TX hospitals for each different claim types
grpby['Hospital', 'Claim Type', 'Percent Spending Hospital'].max()
```
